```python
from pycaret.datasets import get_data
a = get_data('insurance')
```

```python
from pycaret.regression import *
```

```python
a = setup(a, target = 'charges',
           normalize = True,
           polynomial_features = True, trigonometry_features = True, feature_interaction=True, 
           bin_numeric_features= ['age', 'bmi'])
```

```python
a[0].columns
```

```python
lr = create_model('lr')
```

```python
plot_model(lr)
```

```python
evaluate_model(lr)
```

```python
save_model(lr, 'ahraz_minor2_prediction')
```

```python
ahraz_minor2_prediction = load_model('ahraz_minor2_prediction')
```
